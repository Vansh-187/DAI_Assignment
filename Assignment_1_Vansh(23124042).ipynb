##Encoding of Categorical Data

```python
for column in cat_col:
  print(column,':',df[column].unique())
```

```python
label_encoding_col = []
ordinal_encoding_col = []
for column in cat_col:
  if len(df[column].unique()) == 2:
    label_encoding_col.append(column)
  else:
    ordinal_encoding_col.append(column)
```

```python
print(label_encoding_col)
print(ordinal_encoding_col)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for column in label_encoding_col:
  df[column] = label_encoder.fit_transform(df[column])
df["Peer_Influence"].replace({"Positive":2, "Neutral":1, "Negative":0}, inplace=True)
df["Distance_from_Home"].replace({"Far":2, "Moderate":1, "Near":0}, inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
for column in ordinal_encoding_col:
  if ((column!= "Peer_Influence") & (column!="Distance_from_Home")):
        df[column] = ordinal_encoder.fit_transform(df[[column]])
```
